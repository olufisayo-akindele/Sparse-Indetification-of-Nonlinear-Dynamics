```python
import matplotlib.pyplot as plt
import numpy as np

# Creating data for 3D surface plot
x = np.linspace(-30, 30, 100)
t = np.linspace(0, 20, 100)
X, T = np.meshgrid(x, t)
U = np.exp(-0.1 * (X - 5 * T)**2) + np.exp(-0.1 * (X + 5 * T)**2)

fig = plt.figure(figsize=(15, 5))

# Plot A
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, T, U, cmap='gray')
ax1.set_title('A')

# Data for Plot B
x = np.linspace(-10, 10, 400)
u = np.exp(-x**2)

# Plot B
ax2 = fig.add_subplot(132)
ax2.plot(x, u, 'r')
ax2.text(5, 0.8, r'$u_t = cu_x$', fontsize=12, color='black')
ax2.set_title('Linear Waves')

# Data for Plot C
u1 = 0.3*np.exp(-(x)**2)  # Shift right
u2 = np.exp(-(x + 7)**2)  # Shift left

# Plot C
ax3 = fig.add_subplot(133)
ax3.plot(x, u1, 'g')
ax3.plot(x, u2, 'b')
ax3.text(5, 0.8, r'$u_t = 6uu_x + u_{xxx}$', fontsize=12, color='black')
ax3.set_title('Non-Linear Waves')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some sample data
np.random.seed(0)
x = 2.5 * np.random.randn(100) + 1.5   # Array of 100 values with mean = 1.5, stddev = 2.5
res = 0.5 * np.random.randn(100)       # Generate 100 residual terms
y = 2 + 0.3 * x + res                  # Actual values of Y

# Reshape x to be a 2D array
x = x.reshape(-1, 1)

# Create a linear regression model
model = LinearRegression()
model.fit(x, y)

# Calculate the predictions
predictions = model.predict(x)

# Plotting
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, predictions, color='red', label='Regression Line')

# Optionally, add residual lines
for i in range(len(x)):
    plt.plot((x[i], x[i]), (y[i], predictions[i]), 'gray', linestyle='--')

plt.title('Linear Regression Fit')
plt.xlabel('Independent variable (X)')
plt.ylabel('Dependent variable (Y)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from skimage.restoration import denoise_tv_chambolle

# Generate noisy data
np.random.seed(0)
t = np.linspace(0, 2 * np.pi, 500)
true_signal = np.sin(t)
noise = np.random.normal(0, 0.2, t.shape)
noisy_signal = true_signal + noise

# Plot the noisy data
plt.figure(figsize=(12, 6))
plt.plot(t, noisy_signal, label='Noisy Signal', color='red')
plt.plot(t, true_signal, label='True Signal', color='blue', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noisy Data')
plt.legend()
plt.grid(True)
plt.show()

# Denoise the data using Total Variation denoising
denoised_signal_tv = denoise_tv_chambolle(noisy_signal, weight=0.1)

# Denoise the data using Savitzky-Golay filter
window_length = 51  # Choose a window length: must be odd
poly_order = 3  # Polynomial order
denoised_signal_sg = savgol_filter(noisy_signal, window_length, poly_order)

# Plot the denoised data using Total Variation
plt.figure(figsize=(12, 6))
plt.plot(t, denoised_signal_tv, label='TV Denoised Signal', color='green')
plt.plot(t, true_signal, label='True Signal', color='blue', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Denoised Data using Total Variation')
plt.legend()
plt.grid(True)
plt.show()

# Plot the denoised data using Savitzky-Golay
plt.figure(figsize=(12, 6))
plt.plot(t, denoised_signal_sg, label='SG Denoised Signal', color='purple')
plt.plot(t, true_signal, label='True Signal', color='blue', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Denoised Data using Savitzky-Golay Filter')
plt.legend()
plt.grid(True)
plt.show()
```
